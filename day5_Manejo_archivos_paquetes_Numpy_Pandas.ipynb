## Modos en archivos


| Modo | Descripción |
| ---- | ---- |
| r | Abre un archivo para leer. (predeterminado) |
| w | Abre un archivo para escribir. Crea un nuevo archivo si no existe o trunca el archivo si existe. El |
| x | Abre un archivo para creación exclusiva. Si el archivo ya existe, la operación falla. El |
| a | Abre un archivo para agregarlo al final del archivo sin truncarlo. Crea un nuevo archivo si no existe. El |
| t | Se abre en modo texto. (predeterminado) |
| b | Se abre en modo binario. El |
| + | Abre un archivo para actualizar (lectura y escritura) |

```
# Creacion y escritura de un archivo
f = open("list.txt", "w+")

for i in range(10):
  cadena = "linea "+str(i) + "\n"
  f.write(cadena)
f.close()

# Mostrar lo escrito arriba
f = open("list.txt", "r")
for line in f.readlines():
  print(line)
```

```
# Adicionando elementos al final del archivo
f = open("list.txt", 'a+')
for i in range(25, 32):
  cadena = str(i) + "\n"
  f.write(cadena)

f.close()

# Leyendo el archivo list.txt
f = open("list.txt", "r")
for line in f.readlines():
  print(line, end="")
```

## Adicionando una columna "risk" riesgo por zona 

```
def getRisk(neighborhood):
    """Funcion que obtiene el riesgo de una zona. [alto, moderado, bajo, muy bajo]"""

    diccNeighborhoods = {'Achumani' : 'alto', 'Alto Irpavi' : 'bajo', 'Alto Obrajes': 'moderado',  'Aranjuez': 'moderado',  'Auquisamaña': 'bajo',  'Bella Vista': 'moderado',  'Bolognia': 'bajo', 
                     'Calacoto': 'bajo',  'Centro': 'bajo', 'Chasquipampa': 'bajo',  'Cota Cota': 'moderado',  'El Pedregal': 'alto', 'El Tejar ':'moderado',  'Irpavi':'moderado',  'Koani':'bajo', 
                     'La Florida':'moderado',  'Llojeta':'alto',  'Los Pinos':'moderado',  'Mallasa':'alto',  'Mallasilla':'alto',  'Mecapaca':'alto',  'Miraflores':'muy bajo',  'Obrajes':'bajo',
                     'Ovejuyo':'moderado',  'Pampahasi':'bajo', 'Pura Pura':'moderado', 'San Alberto':'alto',  'San Jorge':'muy bajo',  'San Miguel':'bajo', 'San Pedro':'moderado',  'Seguencoma':'bajo',  
                     'Sopocachi':'moderado', 'Tembladerani':'moderado', 'Urb. Autopista':'bajo', 'Villa Copacabana':'muy bajo', 'Villa Fatima':'bajo', 'Villa San Antonio':'alto' }

    for key, value in diccNeighborhoods.items():
        if key.lower().strip() == neighborhood.lower().strip():
            return value
    return 'S/N'
    
```

```
# funcion lambda
dfd['riesgo'] = dfd.apply(lambda row: getRisk(row['neighborhood']), axis=1)
```

```
dfd.head()
```

```
# Verificando cuantas zonas no tienen riesgo
print(dfd[dfd['riesgo'] == 'S/N'].shape)
```
