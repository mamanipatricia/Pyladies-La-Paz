#  Orden de Precedencia
El orden de precedencia de ejecución de los operadores aritméticos es:
    
- Exponente: **
- Negación: -
- Multiplicación, División, División entera, Módulo: *, /, //, %
- Suma, Resta: +, -

**Nota:**
 Se puede omitir este orden de precedencia de ejecución de los operadores aritméticos usando paréntesis () anidados entre cada nivel calculo

```python
2**1/12
```

```python
2**(1/12)
```

### Operadores lógicos

|    expr1| expr2 | Resultado  |

| True     | True     | True     |

|    expr1| expr2 | Resultado  |
| -------- | -------- | -------- |

### IF

```python
a = 10
b = 2

if a > b:
    print("a es mayor que b")
```

```python
# If
edad = 20
if edad > 18:
    print("usted ya es un adulto mayor responsable por favor madure")
# If else
edad = 10
if edad > 18:
    print("usted ya es un adulto mayor responsable por favor madure")
else :
    print("usted es menor de edad")
```

```python
# elif

number = -2

if number > 0:
    print("el numero es positivo")
elif number < 0:
    print("el numero es negativo")
elif number == 0:
    print("el numero es 0")
```

```python
nombre_a = "Juan"
nombre_b = "Carlos"

if nombre_a == nombre_b :
    print("los nombres son iguales")
else:
    print("los nombres son distintos")

# Dias de la semana  
# primera f
dia_actual = "Lunes"

if dia_actual == "Sabado" :
    print("Fin de Semana")
elif dia_actual == "Domingo":
    print("Fin de Semana")
else:
    print("Dia de la semana")

# Segunda forma, optimizando codigo
dia_actual = "Sabado"
if dia_actual == "Sabado" or dia_actual == "Domingo" :
    print("Fin de Semana")
else:
    print("Dia de la semana")
```

```python
# break

for i in range(10):
    if i == 5:
        print('ingrese al 5 y adios')
        break
    print('i => ' + str(i))

# 0 1 2 3 
```

```python
# continue
for i in range(10):
    if i == 5:
        print('ingrea al 5, La siguiente linea ya no se ejecutara')
        continue
    print('i => ' + str(i))
```

```python
# pass
for i in range(10):
    if i == 5:
        print('ingrese al cinco e hice algo')
        pass
    print('i => ' + str(i))
```
